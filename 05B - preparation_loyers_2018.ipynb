```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_app = pd.read_csv('./databases/loyers 2018/indicateurs-loyers-appartements.csv', sep = ";", encoding = 'ansi')
df_app['Type local'] = "Appartement"
df_app.head()
```

```python
df_mai = pd.read_csv('./databases/loyers 2018/indicateurs-loyers-maisons.csv', sep = ";", encoding = 'ansi')
df_mai['Type local'] = "Maison"
df_mai = df_mai.rename(columns = {'Nbobs_maille': 'NBobs_maille'})
df_mai.head()
```

```python
df = pd.concat([df_app, df_mai], axis = 0, ignore_index=True)
df.shape
```

```python
df.head()
```

```python
df[df.columns[7:11]] = df[df.columns[7:11]].apply(lambda x: x.str.replace(',','.'))
df[df.columns[7:11]] = df[df.columns[7:11]].astype(float)
df.head()
```

```python
df.info()
```

```python
df[df.duplicated(subset = ['INSEE', 'Type local']) == True].count()
```

```python
df['INSEE'] = df['INSEE'].str.zfill(5)
df.head()
```

```python
df = df.drop(columns = {'id_zone', 'EPCI'})
```

```python
df.head()
```

```python
df.to_csv('05 - loyers_2018_base_full.csv', encoding = 'UTF-8', index = False)
```

```python
df = df[['INSEE', 'Type local', 'loypredm2']]
df.head()
```

```python
df2 = pd.pivot_table(df, values = 'loypredm2', index = ['INSEE'], columns = ['Type local'], aggfunc = np.mean).reset_index()
df2 = df2.rename(columns = {'Appartement': 'loy_appart', 'Maison': 'loy_maison'})
df2.head()
```

```python
df2.to_csv('05 - loyers_2018_par_commune_typelocal.csv', encoding = 'UTF-8', index = False)
```
