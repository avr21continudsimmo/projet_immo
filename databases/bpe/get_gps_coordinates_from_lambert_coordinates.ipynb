```python
import pandas as pd
bpe = pd.read_csv('databases/bpe19_ensemble_xy.csv', sep = ";", dtype = {'DEP': str, 'DEPCOM': str})
bpe = bpe[-bpe['LAMBERT_X'].isna()]
bpe = bpe.drop(columns={'REG'})
```

```python
bpe.head()
```

```python
from pyproj import Proj, transform
```

```python
inProj = Proj(init='epsg:2154')
outProj = Proj(init='epsg:4326')
bpe['lat'] = ""
bpe['lon'] = ""

for row in bpe.index:
    lat, lon = transform(inProj, outProj, bpe['LAMBERT_X'][row], bpe['LAMBERT_Y'][row])
    bpe['lat'][row] = lat
    bpe['lon'][row] = lon
```
