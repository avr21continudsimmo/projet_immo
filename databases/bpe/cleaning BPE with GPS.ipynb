```python
import pandas as pd
import numpy as np
```

```python
df_bpe = pd.read_csv('bpe_with_gps.csv')
df_bpe.head()
```

```python
df_bpe['vraie_lat'] = df_bpe['lon']
df_bpe['vraie_lon'] = df_bpe['lat']
df_bpe.head()
```

```python
df_bpe = df_bpe.drop(columns = {'lat', 'lon'})
df_bpe.head()
```

```python
df_bpe = df_bpe.rename(columns = {'vraie_lat': 'lat', 'vraie_lon': 'lon'})
df_bpe.head()
```

```python
df_bpe['code_iris'] = df_bpe['DCIRIS'].str.replace('_', '')
df_bpe.head()
```

```python
df_bpe = df_bpe.rename(columns = {'code_iris': 'code_iris_clean'})
```

```python
df_bpe['code_iris_clean'] = df_bpe['code_iris_clean'].apply(lambda x: x + "0000" if len(x) == 5 else x)
```

```python
df_bpe['code_iris_clean'].value_counts()
```

```python
df_bpe.head()
```

```python
df_bpe = df_bpe.drop(columns = {'Unnamed: 0', 'LAMBERT_X', 'LAMBERT_Y', 'DCIRIS'})
df_bpe.head()
```

```python
df_bpe.to_csv('bpe_with_gps_2.csv', index = False)
```
