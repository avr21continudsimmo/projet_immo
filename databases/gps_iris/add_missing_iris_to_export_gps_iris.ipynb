```python
import pandas as pd
```

```python
df = pd.read_csv('export_gps_iris_ALL.csv', dtype = {'Code postal 5 chiffres': str, 'Code commune INSEE': str,
                                             'code_iris': str})
df.head()
```

```python
missing_iris = pd.read_csv('iris_missing_completed.csv')
missing_iris.head()
```

```python
missing_iris = missing_iris.drop(columns = {'Unnamed: 0', 'index', 'Commune_Lat', 'Commune_Lon', 'lon', 'lat', 'Code departement',
                            'Code postal 5 chiffres', 'Code commune INSEE'})
missing_iris.head()
```

```python
full_export = df.merge(missing_iris, on = "Adresse", how = "left")
```

```python
full_export['code_iris_clean'] = full_export['code_iris_clean'].where(full_export['code_iris'].isna(), full_export['code_iris'])
```

```python
full_export[full_export['code_iris_clean'].isna()].head(30)
```

```python
full_export[full_export['code_iris'].isna()].head()
```

```python
full_export = full_export.drop(columns = {'code_iris'})
full_export.head()
```

```python
full_export['code_iris_clean'].isna().sum()
```

```python
full_export.to_csv('export_gps_iris_ALL.csv')
```
