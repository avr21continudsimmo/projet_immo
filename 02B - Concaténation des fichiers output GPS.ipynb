# Objectif du script :
- Concaténer les exports contenant les coordonnées GPS de la base adresse (des exports sont régulièrement effectués lors de l'interrogation des coordonnées GPS car l'API peut planter)

```python
import pandas as pd
import os
from time import time
```

```python
year = '2020'
```

```python
# Récupération de la liste des fichiers export_adresses.csv

file_list = [ f for f in os.listdir('./get_gps/.') if 
    (os.path.isfile(os.path.join('./get_gps/.',f)) & f.lower().endswith(year + '.csv')) ]
print(file_list)
```

```python
adresses = pd.DataFrame(data=[])
nb_files = 0

heure_debut = time()

for f in file_list:
    df = pd.read_csv('./get_gps/' + f)
    adresses = pd.concat([adresses, df])
    nb_files += 1
    print(f)

heure_fin = time()
temps = heure_fin - heure_debut

print("\n")
print("Task achieved in", temps, "s")
print("Nb files", nb_files)
```

```python
import numpy as np
adresses['Code postal'] = adresses['Code postal'].fillna(-1).astype(int).astype(str).replace('-1', np.nan)
adresses['Code postal 5 chiffres'] = adresses['Code postal'].apply(lambda x: x.zfill(5) if len(str(x)) == 4 else x).astype(str)
adresses.head()
```

```python
adresses['Code departement'] = adresses['Code postal 5 chiffres'].str[0:2].astype(str)
adresses.head()
```
