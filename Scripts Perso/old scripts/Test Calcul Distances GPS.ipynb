```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('databases/export_gps_iris_0-10000.csv')
df.head()
```

```python
df_bpe = pd.read_csv('databases/bpe_with_gps_2.csv')
df_bpe.head()
```

```python
df_bpe['gps'] = df_bpe[['lat', 'lon']].apply(tuple, axis=1)
df_bpe.head()
```

```python
from geopy.distance import geodesic

df_test = df[df.index <= 5]
df_test['gps'] = df_test[['lat', 'lon']].apply(tuple, axis=1)
df_test.head()
```

```python
df_test['gps'][0]
```

```python
df_bpe['dist'] = df_bpe['gps'].apply(lambda x: geodesic(x, df_test['gps'][0]).kilometers)
df_bpe.head()
```
