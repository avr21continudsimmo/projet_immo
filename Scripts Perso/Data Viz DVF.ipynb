```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize = (20,10))
plt.scatter(df['Nature mutation'], df['Valeur fonciere']);
```

```python
df.boxplot(column= 'Valeur fonciere', by='Nature mutation', figsize= (20,10));
```

```python
df[df['Nature mutation'] == "Vente"].boxplot(column = 'Valeur fonciere', by='Type local', figsize= (20,10));
```

```python
prix_nature = df.groupby('Nature mutation')['Valeur fonciere'].mean().reset_index()
plt.figure(figsize = (20,10))
plt.bar(prix_nature['Nature mutation'], prix_nature['Valeur fonciere']);
```

```python
prix_typelocal = df.groupby('Type local')['Valeur fonciere'].mean().reset_index()
plt.figure(figsize = (20,10))
plt.bar(prix_typelocal['Type local'], prix_typelocal['Valeur fonciere']);
```

```python
nb_nature = df.groupby('Nature mutation')['Valeur fonciere'].count().reset_index()
plt.figure(figsize = (20,10))
plt.bar(nb_nature['Nature mutation'], nb_nature['Valeur fonciere']);
```

```python
nb_typelocal = df.groupby('Type local')['Valeur fonciere'].count().reset_index()
plt.figure(figsize = (20,10))
plt.bar(nb_typelocal['Type local'], nb_typelocal['Valeur fonciere']);
```

```python
df[(df['Nature mutation'] == "Vente") & ((df['Type local'] == "Maison") | (df['Type local'] == "Appartement"))].isna().sum()
```

```python
df_selec = df[(df['Nature mutation'] == "Vente") & ((df['Type local'] == "Maison") | (df['Type local'] == "Appartement"))]
df_selec.shape
```

```python
df_selec[df_selec['Nature mutation'] == "Vente"].boxplot(column = 'Valeur fonciere', by='Type local', figsize= (20,10));
```

```python
df_selec['prix_m2'] = df_selec['Valeur fonciere'] / df_selec['Surface reelle bati']
df_selec.head()
```

```python
prix_m2_dep = df_selec.groupby('Code departement')['prix_m2'].mean().reset_index()
prix_m2_dep['Code departement'] = prix_m2_dep['Code departement'].astype('str')
```

```python
import seaborn as sns

plt.figure(figsize=(20, 40))
ax = sns.catplot(data = prix_m2_dep, x = 'prix_m2', y = 'Code departement');
```
