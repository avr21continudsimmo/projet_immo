```python
import pandas as pd
```

```python
df = pd.read_csv('dvf_out_2020.csv', dtype = {'Code postal 5 chiffres': str, 'Code commune INSEE': str,
                                             'code_iris_clean': str})
df.head()
```

```python
missing_iris = pd.read_csv('databases/gps_iris/iris_missing_completed.csv')
missing_iris.head()
```

```python
missing_iris = missing_iris.drop(columns = {'Unnamed: 0', 'index', 'Commune_Lat', 'Commune_Lon', 'lon', 'lat', 'Code departement',
                            'Code postal 5 chiffres', 'Code commune INSEE'})
missing_iris.head()
```

```python
dvf = df.merge(missing_iris, on = "Adresse", how = "left")
```

```python
dvf['code_iris_clean'] = dvf['code_iris_clean_y'].where(dvf['code_iris_clean_x'].isna(), dvf['code_iris_clean_x'])
```

```python
dvf[dvf['code_iris_clean_y'].isna()].head()
```

```python
dvf[dvf['code_iris_clean_x'].isna()].head()
```

```python
dvf = dvf.drop(columns = {'code_iris_clean_x', 'code_iris_clean_y'})
dvf.head()
```

```python
dvf.to_csv('dvf_out_2020_cleaniris.csv')
```
